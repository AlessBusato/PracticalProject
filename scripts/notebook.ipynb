```python
from network import NestedSpectralPartition
from pathlib import Path
import scipy.io as scio

import pandas as pd
import numpy as np
from scipy import stats
from network import NestedSpectralPartition

basedir = Path(r"C:\Users\busat\Desktop\PP\prac_proj")

pathin = basedir / "data"
pathout = basedir / "nsp"

fc_files = [fc for fc in pathin.iterdir() if fc.suffix.lower() == '.mat']

FC_data = {}
#nsp_data = {}
nsp_balance = {}

for fc_file in fc_files:
    var_name = fc_file.stem

    mat_contents = scio.loadmat(fc_file)

    var_keys = [k for k in mat_contents.keys() if not k.startswith("__")]
    if not var_keys:
            raise ValueError(f"No variable found in {fc_file.name}") 

    FC_matrix = mat_contents[var_keys[0]]  # should be 400 x 400 x subjects

    FC_data[var_name] = FC_matrix
```

```python
# reshape fc from (400,400,100) to (100,400,400)
fc = FC_data['FC_wm']
fc = np.transpose(fc, (2, 0, 1))
fc_master = np.mean(fc, axis=0)
print(fc_master.shape)
print(fc.shape)
```

```python
nsp = NestedSpectralPartition(fc[:,:,:])
nsp.estimate()
nsp.calibrate(fc_master)
```
